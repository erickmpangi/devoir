```python
import matplotlib.pyplot as plt
import seaborn as sb

plt.figure(figsize=(20, 6))  # Largeur augmentée
sb.scatterplot(x=datos['Country Code'], y=datos['2023'], hue=datos['Indicator Name'])
plt.xticks(rotation=45, ha='right')  # Rotation des labels x
plt.tight_layout()  # Ajuste automatiquement la disposition
plt.show()
```

```python
import numpy as np 
```

```python
# Filtrer les pays avec des valeurs manquantes
datos = datos[datos['2020'].notna()]
# Créer un graphique à barres
plt.figure(figsize=(16, 8))
sb.barplot(x=datos['Country Name'], y=datos['2020'], hue=datos['Indicator Name'])
plt.xticks(rotation=90, ha='right')
# Définir des graduations régulières sur l'axe y
y_min = datos['2020'].min()
y_max = datos['2020'].max()
plt.yticks(np.linspace(y_min, y_max, num=10))  # 10 graduations régulières
plt.tight_layout()
plt.show()
```

```python
datos_na = datos.select_dtypes(include=['float64', 'int64'])  # ou include='number'
sb.set(rc={'figure.figsize': (25, 8)})
sb.heatmap(datos_na.corr(), annot=True, cmap="coolwarm")
```

```python
country="Congo, République démocratique du"
data_country = datos[datos['Country Name'] == country]
columns = [col for col in datos.columns if col.strip().isdigit()]
year=list(map(int, columns))

population=data_country[columns].iloc[0]

df_plot = pd.DataFrame({
    'Année': list(map(int, year)),
    'Valeur': population.values
})
```

```python
plt.figure(figsize=(16, 6))
plt.plot(df_plot['Année'], df_plot['Valeur'], marker='o', linestyle='-')
plt.title(f"Évolution de l'indicateur pour {country}")
plt.xlabel("Year")
plt.ylabel("Population")
plt.xticks(year, rotation=45)
#plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filtrer les pays avec des valeurs manquantes# Créer un graphique à barres
plt.figure(figsize=(16, 8))
sb.barplot(x= year, y= population, hue=['Population with access to internet']*len(year))
plt.xticks(rotation=45, ha='right')
# Définir des graduations régulières sur l'axe y
y_min = population.min()
y_max = population.max()
plt.yticks(np.linspace(y_min, y_max, num=10))  # 10 graduations régulières
plt.tight_layout()
plt.show()
```
