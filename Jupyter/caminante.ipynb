```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres
n_steps = 1000  # Nombre d'étapes

# Directions possibles en 2D : ±x, ±y
steps = np.array([
    [1, 0],   # +x
    [-1, 0],  # -x
    [0, 1],   # +y
    [0, -1],  # -y
])

# Choix aléatoire de direction à chaque étape
moves = steps[np.random.randint(0, 4, size=n_steps)]

# Calcul cumulatif des positions
positions = np.cumsum(moves, axis=0)
positions = np.vstack(([[0, 0]], positions))  # Ajouter le point de départ

# Séparation des coordonnées
x, y = positions[:, 0], positions[:, 1]

# --- Affichage 2D ---
plt.figure(figsize=(16, 8))
plt.plot(x, y, color='black', linewidth=1.8)

# Marquer le départ et l'arrivée
plt.scatter(x[0], y[0], color='green', s=50, label='Departur')
plt.scatter(x[-1], y[-1], color='red', s=50, label='Arrival')

# Titre et axes
plt.title(f'Random Walk ({n_steps} steps)', fontsize=12)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import requis pour les axes 3D

# Paramètres
n_steps = 1000  # Nombre d'étapes

# Directions possibles en 3D : ±x, ±y, ±z
steps = np.array([
    [1, 0, 0],   # +x
    [-1, 0, 0],  # -x
    [0, 1, 0],   # +y
    [0, -1, 0],  # -y
    [0, 0, 1],   # +z
    [0, 0, -1],  # -z
])

# Choix aléatoire de directions
moves = steps[np.random.randint(0, 6, size=n_steps)]

# Calcul des positions cumulées
positions = np.cumsum(moves, axis=0)
positions = np.vstack(([[0, 0, 0]], positions))  # Inclure le point de départ

# Séparation des coordonnées
x, y, z = positions[:, 0], positions[:, 1], positions[:, 2]

# --- Affichage 3D ---
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z, color='black', linewidth=1.5, label='Parcours aléatoire')
ax.scatter(x[0], y[0], z[0], color='green', s=50, label='Départ')
ax.scatter(x[-1], y[-1], z[-1], color='red', s=50, label='Arrivée')

ax.set_title(f'Marche aléatoire 3D ({n_steps} étapes)', fontsize=12)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.tight_layout()
plt.show()
```
