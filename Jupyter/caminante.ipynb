```python
import numpy as np
import matplotlib.pyplot as plt

# Paramtres
n_steps = 2000000  # Nombre d'tapes

# Directions possibles en 2D : x, y
steps = np.array([
    [1, 0],   # +x
    [-1, 0],  # -x
    [0, 1],   # +y
    [0, -1],  # -y
])

# Choix alatoire de direction  chaque tape
moves = steps[np.random.randint(0, 4, size=n_steps)]

# Calcul cumulatif des positions
positions = np.cumsum(moves, axis=0)
positions = np.vstack(([[0, 0]], positions))  # Ajouter le point de dpart

# Sparation des coordonnes
x, y = positions[:, 0], positions[:, 1]

# --- Affichage 2D ---
plt.figure(figsize=(16, 8))
plt.plot(x, y, color='black', linewidth=1.8)

# Marquer le dpart et l'arrive
plt.scatter(x[0], y[0], color='green', s=50, label='Departur')
plt.scatter(x[-1], y[-1], color='red', s=50, label='Arrival')

# Titre et axes
plt.title(f'Random Walk ({n_steps} steps)', fontsize=12)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.legend()

plt.tight_layout()
plt.show()
```
