```python
import os
import matplotlib.pyplot as plt
import numpy as np

# Dossier de sortie
output_dir = "/home/erick-mpangi/Documents/exerices-python/graph_algo"
os.makedirs(output_dir, exist_ok=True)

# Définition des fonctions E2
expressions_E2 = [
    ("1. 2^(2^n)", lambda n: 2**(2**n)),
    ("2. 2^(n^2)", lambda n: 2**(n**2)),
    ("3. n^2 log n", lambda n: n**2 * np.log(n)),
    ("4. n", lambda n: n),
    ("5. 2^n * 2", lambda n: 2**n * 2),
]

# Intervalle de n
n_values = np.linspace(1, 6, 200)

# Évaluation et tri des fonctions selon leur valeur finale
evaluated = []
for label, func in expressions_E2:
    try:
        y = func(n_values)
        final_value = y[-1]  # valeur finale pour le tri
        evaluated.append((label, func, y, final_value))
    except Exception as e:
        print(f"Erreur avec {label}: {e}")

# Tri des courbes par ordre de grandeur croissant
evaluated_sorted = sorted(evaluated, key=lambda item: item[3])  # tri par valeur finale

# Tracé du graphe
plt.figure(figsize=(10, 6))

for label, func, y, _ in evaluated_sorted:
    plt.plot(n_values, y, label=label)

plt.title("E2 - Comparison of Function Growth (Sorted)")
plt.xlabel("n")
plt.ylabel("f(n)")
plt.grid(True)
plt.legend(loc="upper left")
plt.tight_layout()

# Sauvegarde
filename = os.path.join(output_dir, "E2_sorted_functions.png")
plt.savefig(filename)
plt.show()

print(f"Graphe trié sauvegardé dans : {filename}")
```
