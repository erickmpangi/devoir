```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 500)  # temps de 0 à 1 seconde
f1 = np.sin(2 * np.pi * t)       # fréquence de base
f2 = 0.5 * np.sin(4 * np.pi * t) # harmonique

signal = f1 + f2  # Signal combiné

plt.plot(t, signal, label="Signal composé")
plt.plot(t, f1, '--', label="sin(2πt)")
plt.plot(t, f2, '--', label="0.5·sin(4πt)")
plt.legend()
plt.title("Décomposition d'un signal en sinusoïdes")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Données d'exemple
X = np.array([[1], [2], [3], [4], [5]])
Y = np.array([2, 4, 5, 4, 5])

# Modèle
model = LinearRegression()
model.fit(X, Y)

# Prédictions
Y_pred = model.predict(X)

# Affichage
plt.scatter(X, Y, color='blue')
plt.plot(X, Y_pred, color='red')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.show()

# Performance
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("MSE:", mean_squared_error(Y, Y_pred))
print("R^2:", r2_score(Y, Y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_steps = 1000
mu, sigma = 0.1, 0.5  # Drift and volatility

# Simulation
steps = np.random.normal(mu, sigma, n_steps)
random_walk = np.cumsum(steps)

# Visualization
plt.figure(figsize=(10, 6))
plt.plot(random_walk, label=f"Random walk with drift μ={mu}, σ={sigma}")
plt.title("Financial Random Walk Simulation")
plt.xlabel("Time (days)")
plt.ylabel("Asset Price")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
from statsmodels.tsa.stattools import adfuller

# Exemple de données : simulation de prix d'un actif
np.random.seed(42)  # Pour des résultats reproductibles
stock_prices = np.cumsum(np.random.normal(0, 1, 100))  # Random walk

# Test de stationnarité avec le test d'ADF
result = adfuller(stock_prices)
p_value = result[1]  # Si p > 0.05, on ne rejette pas H0 (random walk)

print("P-value:", p_value)
if p_value > 0.05:
    print("La série est non stationnaire (random walk).")
else:
    print("La série est stationnaire.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configure plot aesthetics
sns.set_palette("husl")
plt.rcParams['figure.facecolor'] = 'white'

# Data
neighborhoods = ['Q1', 'Q2', 'Q3']
population = np.array([3000, 4000, 3000])
infected = np.array([300, 800, 1200])

# Metrics
infection_rates = infected / population
relative_risk = infection_rates / infection_rates[0]

# Plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Infection Rates
ax1.bar(neighborhoods, infection_rates, width=0.6)
ax1.set_title('Infection Rates by Neighborhood', fontweight='bold')
ax1.set_ylabel('Infection Rate')
ax1.set_ylim(0, 0.5)
for i, rate in enumerate(infection_rates):
    ax1.text(i, rate + 0.02, f'{rate:.0%}', ha='center')

# Plot 2: Relative Risk
ax2.bar(neighborhoods, relative_risk, width=0.6, color='orange')
ax2.set_title('Relative Risk (vs Q1)', fontweight='bold')
ax2.axhline(1, color='gray', linestyle='--')
for i, risk in enumerate(relative_risk):
    ax2.text(i, risk + 0.1, f'{risk:.1f}x', ha='center')

plt.tight_layout()
plt.savefig('epidemic_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Prompt the user for input
number = int(input("Enter an integer: "))

# Convert the number to binary string (remove '0b' prefix)
binary = bin(number)[2:]

# States represent the remainder modulo 3: 0, 1, 2
state = 0

# Transition table: (current_state, input_bit) -> next_state
transition = {
    (0, '0'): 0,
    (0, '1'): 1,
    (1, '0'): 2,
    (1, '1'): 0,
    (2, '0'): 1,
    (2, '1'): 2
}

# Process each bit in the binary representation
for bit in binary:
    state = transition[(state, bit)]

# If we end up in state 0, the number is divisible by 3
if state == 0:
    print(f"{number} is divisible by 3.")
else:
    print(f"{number} is NOT divisible by 3.")
```
